# Data Processing


### **STEP-1 Handling Missing values**
We don't want compounds that do not have values for **standard_value** and **canonical_smiles** columns. We create another dataframe **df2** that stores data of only those rows for which there is no missing value ie it is NOT NA.(Missing value is referred as NA).

**Standard_value** : The potency of the compound. Lower the value, greater the potency.

**Canonical_smiles** : Representation of the chemical structure of the compound in SMILES notation.

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

### **STEP-2 Deleting duplicates**

Now, we also do not want compounds that have the same chemical structure, ie canonical_smiles. So, we drop any duplicates.

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

### **STEP-3 DataFrame with only required columns :**

Now, we create another dataframe **df3** that contains only the 3 important columns : molecule_chembl_id, canonical_smiles, standard_value.

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

Now, we save the dataframe in a CSV file.

```python
df3.to_csv('bioactivity_data_preprocessed.csv', index=False)
```

### **STEP-4 Creating Another Column : Bioactivity Class**
We label the compunds as ACTIVE, INACTIVE & INTERMEDIATE based on their standard_value related to IC50.
Lower the standard_value, greater the potency and inhibition effect, greater the drug activity of the compound.

**Active** : standard_value < 1000 nM

**Inactive** : standard_value > 10000 nM

**Intermediate**: 1000 nM < standard_value < 10000 nM

```python
df4 = pd.read_csv('bioactivity_data_preprocessed.csv')
```

```python
bioactivity_threshold = [] #Creating a list using iteration
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

### **STEP-5 Combining New column with existing columns**

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='bioactivity_class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

### **STEP-6 Saving as CSV file**

```python
df5.to_csv('bioactivity_data_fullypreprocessed.csv', index=False)
```

# **Adding Lipinski Descriptors**

Not all chemical compounds can be used as drugs. Based on **ADME** (also known as the **pharmokinetic** profile)- *Absorption, Distribution, Metabolism and Excretion*, **Christopher Lipinski**, a scientist at Pfizer, came up with a set of thumb rules. This Rule-of-five (as all values are multiples of 5) examines the druglikeness of a chemical compound.


**The Lipinski's Rule stated the following:**



Molecular weight < 500 Dalton

Octanol-water partition coefficient (LogP) < 5

Hydrogen bond donors < 5

Hydrogen bond acceptors < 10

## **Importing required libraries and modules**

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
df = pd.read_csv('bioactivity_data_fullypreprocessed.csv')
df
```

## **Calculate descriptors**

```python
# defining a function lipinski to calculate all 4 types of descriptors
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

Applying this function to our data (to the `canonical_smiles` column) to calculate the molecular descriptors.

```python
df_lipinski = lipinski(df.canonical_smiles)
df_lipinski
```

## **Combining Initial dataframe df with the Lipinksi dataframe**

```python
df_combined = pd.concat([df,df_lipinski], axis=1)
df_combined
```

## **Convert IC50 to pIC50**
To allow **IC50** standard_value data to be more uniformly distributed for machine learning, we will convert **IC50** to the negative logarithmic scale which is **-log10(IC50)**.

This custom function pIC50() will accept a DataFrame as input and will:
* Take the IC50 values from the ``standard_value`` column and converts it from nM to M by multiplying the value by 10$^{-9}$
* Take the molar value and apply -log10
* Delete the ``standard_value`` column and create a new ``pIC50`` column

However, we will first apply a custom norm_value() function so that the values in the standard_value column are normalized. This essentially caps the value at 10,00,00,000 as values greater than 10,00,00,000 give negative logarithmic value which is not desired.

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop(columns='standard_value')

    return x
```

```python
df_norm = norm_value(df_combined)
df_norm
```

```python
# now defining the pIC50 function
def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop(columns='standard_value_norm')

    return x
```

```python
df_pIC50 = pIC50(df_norm)
df_pIC50
```
